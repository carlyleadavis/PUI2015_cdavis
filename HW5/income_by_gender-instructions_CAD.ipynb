##to see how the male and female income relate to each other plot the Total Median Income (from any source of revenue) of females vs the Total Median Income of males by race (one data point per race)

##also plot the 1-1 line: the diagonal that corresponds to men and women getting the same pay


```python
categories = ['all', 'white', 'black', 'hispanic', 'asian']
color_key = {'all':'k', 'white':'c', 'black':'b', 'hispanic':'g', 'asian':'r'}

#creating the arrays that we will use for our regressions - median income by race for males and females
male_graph1 = []
female_graph1 = []

for race in categories:
    male_append = allmales[race]['Median income'][1]
    female_append = allfemales[race]['Median income'][1]
    male_graph1.append(male_append)
    female_graph1.append(female_append)
```

```python
pl.figure(figsize = (8,8))
for race in categories:
    pl.scatter(allmales[race]['Median income'][1], allfemales[race]['Median income'][1]
               , color = color_key[race], label= race)
    
pl.xlabel('Median Income - Male (USD)')
pl.ylabel('Median Income - Female (USD)')
pl.title('Median Income, Male vs. Female')
pl.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
pl.xlim(0, 45000)
pl.ylim(0, 45000)
pl.plot([0, 42000], [0, 42000], '--', label = '1-1 Line')
pl.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

**The above plot shows us the earning discrepancy by gender.  If all of the points fell along the 1-1 line, we would have equal earning between males and females.  Since all of the point fall below the line, we clearly see that males are earning more than their female counterparts for every race.**

#Now fit it with a line!

```python
import statsmodels.api as sm

model = sm.OLS(female_graph1, male_graph1).fit()
print (model.summary())
print(model.params)
```

##it turns out that there is an analytical solution for the OLS regression it is coded below. please try to read the code and understand the formula! 


```python
#this is the analytical best regression line. it can be proved (ask me about it if you want!)

cor = stats.spearmanr(male_graph1, female_graph1)
beta1=cor[0]*np.std(female_graph1)/np.std(male_graph1)
beta0=np.mean(female_graph1)-beta1*np.mean(male_graph1)
```

```python
#define function line so that we can input our regression stats and graph a fit line easily
def line (x, b1, b0):
    return b1*x + b0
```

##and plot the line corresponding to the analytical regression to compare it with the result of your numerical one

```python

myseed = np.random.seed(1)
sample = np.arange (0,45000, 1)


model_line = line(sample, model.params[0], 0)
calculated_line = line(sample, beta1, beta0)


pl.figure(figsize = (8,8))
for race in categories:
    pl.scatter(allmales[race]['Median income'][1], allfemales[race]['Median income'][1]
               , color = color_key[race], label= race)
    
pl.xlabel('Median Income - Male (USD)')
pl.ylabel('Median Income - Female (USD)')
pl.title('Median Income, Male vs. Female')
pl.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
pl.xlim(0, 45000)
pl.ylim(0, 45000)
pl.plot([0, 42000], [0, 42000], '--', label = '1-1 Line')


pl.plot (model_line, 'k-', color = 'black', label='OLS Regression Line')
pl.plot (calculated_line, 'k-', color = 'green', label='Manual Regression Line')
pl.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```


**The above plot shows us the earning discrepancy by gender. If all of the points fell along the 1-1 line, we would have equal earning between males and females. Since all of the point fall below the line, we clearly see that males are earning more than their female counterparts for every race.  Also on this plot are our stats models regression line, and our manually calculated regression line, which are better indicators of the relationship between male and female median income by race than the 1-1 Line.**

to unravel the data and concatenate them back together in a single array is a little complex (there may be easier way, this is just wha i came up with. if you have something better please share!!)
here ar a few lines of code that should do the trick. i used nu.concatenate to put together arrays (it takes a list of arrays) and np.flatten to make the array 1 dimensional after merging. i also masked the nan values. i used the numpy ma (masled array) package. ever array is associated by this package to a mask array of Bools [True....False...] of the same size. the 

```python
all_males=np.ma.array(np.concatenate([allmales['all']['Median income'][:].values,
                                      allmales['white']['Median income'][:].values,
                                      allmales['black']['Median income'][:].values,
                                      allmales['hispanic']['Median income'][:].values,
                                      allmales['asian']['Median income'][:].values]).flatten())

all_females=np.ma.array(np.concatenate([allfemales['all']['Median income'][:].values,
                                        allfemales['white']['Median income'][:].values,
                                        allfemales['black']['Median income'][:].values,
                                        allfemales['hispanic']['Median income'][:].values,
                                        allfemales['asian']['Median income'][:].values]).flatten())

#I am gonna create the mask int he next line. 
#the mask is True where either male or female value is Nan. 
#Notice that True is really just >1, and False really 0. 
#so i can add the male and female mask and the resulting mask will be 
#False where both arrays are not Nan, and true otherwise. 

allmask = [np.isnan(all_males) + np.isnan(all_females)]

all_females.mask = allmask
all_males.mask = allmask
```

```python
model2 = sm.OLS(~np.isnan(all_females), ~np.isnan(all_males)).fit()
print('OLS Stats Models Regression Params:\t',model.params)


#fit the data with a regression line

model3 = np.polyfit(~np.isnan(all_males), ~np.isnan(all_females), deg=1)
print('NP Polyfit Regression Params:\t\t',model3)
```
